## Removing null-valued rows for certain columns and cleaning some categories

```python
# Checking which categories have null values/how many there are

originaldf.isnull().sum()
```

```python
# Using pandas to update the data by removing rows with NaN for certain categories

originaldf.fillna({'reviews_per_month':0}, inplace=True) #replacing all NaN values
originaldf = originaldf[originaldf['review_scores_value'].notna()] 
originaldf = originaldf[originaldf['beds'].notna()] 
originaldf = originaldf[originaldf['bedrooms'].notna()] 
originaldf = originaldf[originaldf['host_since'].notna()]
originaldf = originaldf[originaldf['bathrooms_text'].notna()] 
#check changes
print(originaldf.reviews_per_month.isnull().sum(), originaldf.beds.isnull().sum(), originaldf.bedrooms.isnull().sum())
```

```python
# changing host_start dates to # of days since they first joined Airbnb

from datetime import datetime

def days_between(d1, d2):
    d1 = datetime.strptime(d1, '%Y-%m-%d')
    d2 = datetime.strptime(d2, '%Y-%m-%d')
    return abs((d2 - d1).days)

originaldf['host_since'] = originaldf['host_since'].map(lambda x: days_between(str(x), "2022-10-21"))
```

```python
# checking if change was implemented properly

originaldf['host_since']
```

```python
# changing beds category into integer value

originaldf['beds'].map(lambda x: int(x))

originaldf['beds']
```

```python
# changing bathrooms_text to integer value

def separate(x):
    text = str(x)
    if "Shared" in text:
        return 1
    elif "Half-bath" in text:
        return 0.5
    elif "Private" in text:
        return 1
    else:
        text2 = float(text.split()[0]) #1 bath
    return(int(text2))

originaldf['bathrooms_text'] = originaldf['bathrooms_text'].map(lambda x: separate(x))

originaldf['bathrooms_text']
```

### Linear Regression

```python
# Re-importing libraries (just in case I run this cell by itself)

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import preprocessing, svm
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

regressiondata = sub_1000[['host_since', 'accommodates', 'review_scores_value', 'bathrooms_text', 
                           'latitude', 'longitude', 'host_listings_count', 'number_of_reviews', 'reviews_per_month', 'price']]
  
# Displaying a few rows of my data
regressiondata
```

```python
# Dropping NaN values because they cannot be used for modeling 
regressiondata = regressiondata.dropna(axis=0) 

# Grabbing dependent and independent variables
x = regressiondata[['host_since', 'accommodates', 'review_scores_value', 'bathrooms_text', 
                           'latitude', 'longitude', 'host_listings_count', 'number_of_reviews', 'reviews_per_month']]
y = regressiondata['price']

regressiondata.dropna(inplace = True)
```

```python
# Training my linear regression model with 80% of data points, and testing regression with 20% of the data points

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
# importing module
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# creating an object of LinearRegression class
LR = LinearRegression()

# fitting the training data
LR.fit(x_train,y_train)

# creating prediction
y_prediction =  LR.predict(x_test)

# print error
print(mean_squared_error(y_test, y_prediction))
```

```python
# Calculating regression score, analogous to R^2 value; 0.4 is enough to say that variation in data is not just due to chance

LR.score(x_test, y_test)
```

```python
# Grabbing array of coefficients, respective to each predictor

coefficient_array = LR.coef_

print(coefficient_array)
```

```python
# Here I am creating a graph that charts each of the coefficients; larger coefficient magnitues influence data more

import numpy as np                                                               
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import math

figure(figsize=(20, 10), dpi=80)
width = 1
# plt.bar(x.columns, np.abs(np.log(np.abs(coefficient_array))), width, align='center')
# horizontal plot of the magnitudes of each coefficient
plt.barh(x.columns, np.abs(np.log(np.abs(coefficient_array))), width, align='center')
```

```python
# Calculating p-values for the regression model coefficients

import statsmodels.api as sm

x_train_w_constant = sm.add_constant(x_train)

model = sm.OLS(y_train, x_train_w_constant).fit()
predictions = model.predict(x_train_w_constant)
print_model = model.summary()

print(print_model)
print(x_train)
```

## Step 3. Clustering

Although it is attractive to segment data into groups such as island (Kauai, Oahu, Maui, and Hawaii), I chose to use k-means clustering because I wanted to see how the data would partition itself. I chose K-means clustering in particular because it scales well to larger datasets and is commonly used. 

```python
# Importing necessary packages

import matplotlib.pyplot as plt
from kneed import KneeLocator
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler
```

```python
# Standardizing features before clustering so they are all on the same scale

scaler = StandardScaler()
scaled_features = scaler.fit_transform(x)

# Trying out k-means clustering for first time

kmeans = KMeans(
init="k-means++",
n_clusters=4,
n_init=10,
max_iter=300,
random_state=42
)
```

```python
kmeans.fit(scaled_features)
```

```python
# Printing some important features 

print(kmeans.cluster_centers_) # centroids for each of the 4 cluster based on 9 features "4 values that dataset revolves around"
# if we wanted to create 4 stereotypes of the data, this is what it is
print(kmeans.n_iter_) # iterations for convergance
print(kmeans.labels_) # which cluster each point belongs to 
print(kmeans.inertia_) # least squares error
```

```python
# setting default arguments for k-means
kmeans_kwargs = {
"init": "random",
"n_init": 10,
"max_iter": 300,
"random_state": 42
}

# generating array of errors to evaluate which cluster size is the best
sse = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, **kmeans_kwargs)
    kmeans.fit(scaled_features)
    sse.append(kmeans.inertia_)
print(sse)

# plotting SSE curve and checking where SSE curve starts to bend (balance between too few and too many clusters)
plt.style.use("fivethirtyeight")
plt.plot(range(1, 11), sse)
plt.xticks(range(1, 11))
plt.xlabel("Number of Clusters")
plt.ylabel("SSE")
plt.show()

# using kneelocator to automatically calculate the elbow 
kl = KneeLocator(range(1, 11), sse, curve="convex", direction="decreasing")
kl.elbow
```

```python
# Silhouette coefficient is a measure of cohesion and separation, quantifies how well data points fit into clusters 
# based on proximity within the cluster and distance between data in different clusters

# A list holds the silhouette coefficients for each k
silhouette_coefficients = []

# Silhoutte coefficients start at 2 clusters (for between cluster differences to be accounted for)
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, **kmeans_kwargs)
    kmeans.fit(scaled_features)
    score = silhouette_score(scaled_features, kmeans.labels_)
    silhouette_coefficients.append(score)
```

```python
# plotting silhouette coefficient graph; silhouette coefficient values range from -1 to 1, where larger numbers indicate 
# samples are closer to their clusters than they are to other clusters

plt.style.use("fivethirtyeight")
plt.plot(range(2, 11), silhouette_coefficients)
plt.xticks(range(2, 11))
plt.xlabel("Number of Clusters")
plt.ylabel("Silhouette Coefficient")
plt.show()
```

```python
# I want to graph my clusters on a 2D plane, so I try to reduce the dimensions of my data through PCA

from sklearn.decomposition import PCA
  
pca = PCA(n_components = 2)

# generating coordinates of eigenvectors
# transforming each row into a 2d vector based on most significant features (two most explainable eigenvectors)

#using pca on scaled features
x_train_pca = pca.fit_transform(scaled_features) 

x_train_pca

kmeans_kwargs = {
"init": "random",
"n_init": 10,
"max_iter": 300,
"random_state": 42
}

# redoing elbow plot to re-do clustering on PCA data
sse = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, **kmeans_kwargs)
    kmeans.fit(x_train_pca)
    sse.append(kmeans.inertia_)
    
print(sse)

plt.style.use("fivethirtyeight")
plt.plot(range(1, 11), sse)
plt.xticks(range(1, 11))
plt.xlabel("Number of Clusters")
plt.ylabel("SSE")
plt.show()

kl = KneeLocator(range(1, 11), sse, curve="convex", direction="decreasing")
kl.elbow
```

```python
# re-doing k-means with pca data 
best_kmeans = KMeans(n_clusters = 4, **kmeans_kwargs)
best_kmeans.fit(x_train_pca)
best_kmeans.labels_

# plotting the left and right coordinates for each 2-d vector
plt.scatter(x_train_pca[:, 0], x_train_pca[:,1], c=best_kmeans.labels_, cmap='rainbow')

#interpretation: coordinates of eigenvectors 
# took 9d, squeezed onto plane that spans 2d
# 4 archetypes of airbnb listings 

plt.scatter(best_kmeans.cluster_centers_[:, 0], best_kmeans.cluster_centers_[:, 1])
```
